```python
#import the packages / dependecies
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np
from keras.utils import to_categorical
import matplotlib.pyplot as plt # Graph
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
# Reads heart disease dataset from file
data = pd.read_csv("TakeHomeExamDatasetHeartDisease.csv")
```

```python
data
```

```python
# Reads appropriate column data for the input and output datasets
X = data[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','slope','ca','thal']].values
Y = data['target'].values
X = X.astype(int)
Y = Y.astype(int)
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X)

# Creates global variables to pass data between classes
model=0
Xtrain=0
Ytrain=0
Xtest=0
Ytest=0
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, random_state=515)
```

```python
Ytrain = to_categorical(Ytrain)
```

```python
# Builds NN to classify heart disease dataset
def buildAndTrainModel(numEpochs) :
    global model #, Xtrain, ytrain, Xtest, ytest
    # Build the NN model - 12 inputs, 12 neurons in hidden layer, 2 outputs
    model = Sequential()
    model.add(Dense(12, input_dim=12))
    model.add(Activation('tanh'))
    model.add(Dense(2)) 
    model.add(Activation('softmax')) 

    # create a Stochastic Gradient Descent (SGD) optimizer with a learning rate of 0.1
    sgd = SGD(lr=0.1)
    
    # build the model NN as described by the above commands
    model.compile(loss='mean_squared_error', optimizer=sgd,metrics = ["accuracy"])
    
    # do a user specified number of iterations of forward and backward propagation to train the network
    # all operations here are done behind the scenes, all updating of weights and
    # calculation of errors
    #Train the model
    model.fit(Xtrain, Ytrain, batch_size=1, epochs=numEpochs)
```

```python
buildAndTrainModel(100)
```

```python
# Used to compute the accuracy of the prediction, so the output of the NN and the integer categories can be compared
def posOfMaxOfTwo(n1, n2) :
    if (n1 >= n2) :
        return 0
    if (n2 >= n1) :
        return 1
```

```python
def predictResults () :
    sum=0
    for i in range(len(Xtest)) :
        results = model.predict_proba(np.array([Xtest[i]]))
        result = posOfMaxOfTwo(results[0][0], results[0][1])
        if (result == posOfMaxOfTwo(Ytest[i][0], Ytest[i][1])):
            sum += 1
        print("Predicted result:",result,"Actual result:",posOfMaxOfTwo(Ytest[i][0], Ytest[i][1]))
    print("Accuracy:",(sum/len(Ytest)) * 100)
```

```python
# reformat the target test data
Ytest = to_categorical(Ytest)
```

```python
predictResults()
```

```python
#Evaluate the model
model.evaluate(
  Xtest,
  Ytest
)
```
